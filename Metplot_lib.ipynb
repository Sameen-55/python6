```python
months = ["Jan","Feb","Mar","Apr","May","Jun"]

product_A = [120,150,170,140,200,220]
product_B = [100,130,160,120,180,210]

expenses = [80,90,110,85,140,150]
import matplotlib.pyplot as plt

plt.plot(months, product_A, marker="o", label="Product A")
plt.plot(months, product_B, marker="o", label="Product B")

plt.title("Product Sales Trend")
plt.xlabel("Months")
plt.ylabel("Sales")

plt.legend()
plt.grid()

plt.show()
```

```python
plt.bar(months, product_A)

plt.title("Product A Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")

plt.show()
```

```python
profit = [(pa + pb - e) for pa, pb, e in zip(product_A, product_B, expenses)]

plt.subplot(2,2,1)
plt.plot(months, product_A)
plt.title("Product A Sales")

plt.subplot(2,2,2)
plt.plot(months, product_B)
plt.title("Product B Sales")

plt.subplot(2,2,3)
plt.bar(months, expenses)
plt.title("Company Expenses")

plt.subplot(2,2,4)
plt.plot(months, profit)
plt.title("Company Profit")

plt.tight_layout()

plt.show()
```
